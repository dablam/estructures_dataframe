# EXERCICI 2
### seleccionar 2 atributs. calcular p-valor. alpha 5% -> rebutja hipòtesi nul·la?

```python
# atribut 1, partits guanyats: W
# atribut 2, gols a favor: GF

guany = uf['W']
gols = uf['GF']

from scipy.stats import ttest_ind

resul = ttest_ind(guany, gols)

print(resul)
```

```python
p_value = .0523838603340441e-53
alpha = 0.05
             
if p_value > alpha:
    print("We DON'T REJECT the null hypothesis H0")
else:
    print("We REJECT the null hypothesis H0")   
```

```python
vis = describe(resul)

print(vis)
```

# EXERCICI 3
### seleccionar 3 atributs. calcular p-valor. alpha 5% -> rebutja hipòtesi nul·la?

```python
# atribut 1, ranking lliga pais: LgRk
# atribut 2, punts: Pts
# atribut 3, màxima puntuació: MP

from scipy.stats import f_oneway

rl = uf['LgRk']
pt = uf['Pts']
mp = uf['MP']

stat, p = f_oneway(rl, pt, mp)
print('stat=%.3f, p=%.3f' % (stat, p))

if p > 0.05:
    print("We DON'T REJECT the null hypothesis H0")
else:
    print("We REJECT the null hypothesis H0")  

if p > 0.05:
 print('Probably the same distribution')
else:
 print('Probably different distributions')
```
