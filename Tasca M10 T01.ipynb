```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
import pandas as pd
```

```python
driver = webdriver.Firefox(service=Service())
driver.get("https://www.duckduckgo.com")
```

```python
url = 'https://www.bolsasymercados.es/bme-exchange/es/Mercados-y-Cotizaciones/Acciones/Mercado-Continuo/Empresas-Cotizadas'
driver.get(url)
```

```python
rows = []
for table in driver.find_elements(by= By.XPATH, value = '//*[contains(@class,"shares-table")]//tr'):
        data = [item.text for item in table.find_elements(by= By.XPATH, value= ".//*[self::td or self::th]")]
        
        #dades per files
        rows.append(data)

em_coti = pd.DataFrame(rows[1:], columns = rows[:])
em_coti

# ara no surt reflexat???
```

```python
class QuotesSpider(scrapy.Spider):
    name = 'resi'
    start_bcn = [
        'https://guia.barcelona.cat/ca/llistat?&pg=search&xout=1&ajax=search&af=code_prop&wtarget=la-meva-barcelona&nfwt=1&code0=0062001014&code1=0062002014012&nr=10&c=00620*&tr=620&nr=10&sort=popularity,desc',
    ]

    def parse(self, response):
        for quote in response.css('resi'):
            yield {
                'residencia': quote.xpath('span/small/text()').get(),
                'text': quote.css('span.text::text').get(),
            }

        next_page = response.css('li.next a::attr("href")').get()
        if next_page is not None:
            yield response.follow(next_page, self.parse)
```
