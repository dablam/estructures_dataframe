```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rc
plt.style.use(['dark_background'])

# descarregar dataframe "preu lloguer per trimestre i barri"

grf = pd.read_csv('preu lloguer per trimestre i barri.csv')

#visualització de les 20 primeres files
grf[:20]
```

# informació del dataframe

```python
#nformació Dtype, columnes i valors

grf.info() 
```

```python
# número de nulls per columna

grf.isnull().sum() 
```

# visualització variable categòrica (districte)

```python
# Nom_Districte dtype object
# Suma de valors per Nom_Districte

barri = grf['Nom_Districte'].value_counts()
barri 
```

```python
# plot valors per districte

barri.plot(figsize = (11,9), kind = 'bar', color = 'g')

#labels del plot

plt.title("VARIABLE CATEGÒRICA")
plt.xlabel("NOM DISTRICTE")
plt.ylabel("VALORS")
```

# visaulització variable numèrica (preu)

```python
# la columna preu té valors '--'
# canvi valors de l'objecte '--' per l'objecte '0'

grf.loc[grf["Preu"] == "--", "Preu"] = 0
grf[10:15]
```

```python
# convertir columna 'Preu', objecte en numèrica, del contrari no es pot plot

grf['Preu'] = pd.to_numeric(grf['Preu']) 

grf.info()
```

```python
# convertir columna 'Preu' en int, per poder plot

grf['Preu'] = grf['Preu'].astype(int)
grf.info()
```

```python
# Preu dtype object
# Suma de valors Preu

preu = grf['Preu'].value_counts()
preu
```

```python
# plot variable numèrica Preu

y = grf['Preu'] # columna per plot 

# definició del plot

y.plot(figsize = (11,9))
plt.plot(y, color = 'r')
plt.grid() # grip per guiar els valors

plt.title("VARIABLE NUMÈRICA")
plt.xlabel("REGISTRES") # valors del 0 al 584 (número total d'entrades per defecte) 
plt.ylabel("PREU")
xlim=[0, 600]
ylim=[0,1600]

# l'eix de les x, al no estar definit, per defecte s'inicia al 0 fins el 584 (número total d'entrades).
# els pics de baixa, són deguts als valos '--' existents en la columna preu i convertita a int 0.
# varietat de valors per 2 registres diferents en variable 'Preu' (preu mitja mensual i preu mitja superfícien)

plt.show()
```

```python
# plot valors per Preu

preu.plot(figsize = (11,9), color = 'g')
plt.grid() # grip per guiar els valors

plt.title("VARIABLE NUMÈRICA")
plt.xlabel("PREU")
plt.ylabel("REGISTRES")
```

# visualització variable numèrica i categòrica (districte + preu)

```python
# selecció dues columnes

vnvc = grf[['Nom_Districte','Preu']]

# agrupació per Nom_Districte

nc = vnvc.groupby('Nom_Districte').count()
nc
```

```python
nc.plot(figsize = (11,9), kind = 'bar', color = 'r')
plt.title("PREU PER DISTRICTE")
plt.xlabel("DISTRICTE")
plt.ylabel("PREU")
```

# visualització dues variables numèriques (trimestre + preu)

```python
# selecció dues columnes

vnvn = grf[['Trimestre','Preu']]

# agrupació per Trimestre

nn = vnvn.groupby(['Trimestre']).value_counts()
nn
```

```python
# informació de les variables a plot

nn.info()
```

```python
nn.plot(figsize = (11,9), color = 'y') 

plt.title("PREU PER TRIMESTRE") 
plt.xlabel("TRIMESTRE")
plt.ylabel("PREU")
xlim=[0, 600]
ylim=[0,1750]

plt.show()
```

# visualització tres variables (barri + trimestre + preu)

```python
# informació columnes seleccionades

dtp = grf[['Nom_Districte', 'Trimestre', 'Preu']]

nd = dtp.groupby(['Nom_Districte', 'Trimestre']).sum()
nd
```

```python
nd.plot(figsize = (11,9), kind='bar', color = 'hotpink') 
plt.title("PREU PER BARRI - TRIMESTRE") 
plt.xlabel("BARRI-TRIMESTRE")
plt.ylabel("PREU")

plt.show()
```

## altres visualitzacions

```python
# selecció lloguer mitjà mensual en la columna 'Preu'

llmm = grf[grf['Lloguer_mitja'] == "Lloguer mitjà mensual (Euros/mes)"]
llmmensual = llmm[['Nom_Districte', 'Preu']]
lmen = llmmensual['Preu'].value_counts()
lmen
```

```python
# selecció lloguer mitjà mensual en la columna 'Preu' 
# (de les 584 files totals, 292 és el lloguer mitjà mensual en la columna 'Preu')

llmm = grf[grf['Lloguer_mitja'] == "Lloguer mitjà mensual (Euros/mes)"]
lmen = llmm[['Nom_Districte', 'Preu']]
lmen 
```

```python
# agrupació de lloguer mitja mensual per Nom_Barri

t_llmm = lmen.groupby('Nom_Districte').sum()
t_llmm
```

```python
# plot lloguer mitja mensual per Nom_Barri

t_llmm.plot(figsize = (15,13), color = 'y')
plt.title("LLOGUER MENSUAL MITJA DISTRICTE")
plt.xlabel("DISTRICTE")
plt.ylabel("PREU")
```

```python
# selecció lloguer mitjà per superficie en la columna 'Preu'
# (de les 584 files totals, 292 és el lloguer mitjà per superfície en la columna 'Preu')

llms = grf[grf['Lloguer_mitja'] == "Lloguer mitjà per superfície (Euros/m2 mes)"]
llmsuperf = llms[['Nom_Districte', 'Preu']]
llmsuperf
```

```python
# agrupació de lloguer mitja superfície per Nom_Barri

t_llms = llmsuperf.groupby('Nom_Districte').sum()
t_llms
```

```python
# plot lloguer mitja superfície per Nom_Barri

t_llms.plot(figsize = (15,13), color = 'y')
plt.title("LLOGUER MITJA SUPERFÍCIE DISTRICTE")
plt.xlabel("DISTRICTE")
plt.ylabel("PREU")
```
